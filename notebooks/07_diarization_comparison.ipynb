# 6. Solo Audio Files Wave Length

```python
import wave
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = "../audios-wav/12-audios-ar-en/6-audios-ar/1_speaker_ar/solo3_ar.wav"
file_path_2 = "../audios-wav/12-audios-ar-en/6-audios-en/1_speaker_en/solo2_en.wav"

def read_waveform(file_path):
    with wave.open(file_path, 'rb') as wf:
        n_frames = wf.getnframes()
        framerate = wf.getframerate()
        sample_width = wf.getsampwidth()
        n_channels = wf.getnchannels()
        raw_frames = wf.readframes(n_frames)
        # Convert bytes to numpy array
        dtype = np.int16 if sample_width == 2 else np.uint8
        waveform = np.frombuffer(raw_frames, dtype=dtype)
        if n_channels > 1:
            waveform = waveform.reshape(-1, n_channels)
        return waveform, framerate

# Read both audio files
waveform_ar, framerate_ar = read_waveform(file_path)
waveform_en, framerate_en = read_waveform(file_path_2)

print(f'Arabic file: {file_path}, Frames: {len(waveform_ar)}, Framerate: {framerate_ar}')
print(f'English file: {file_path_2}, Frames: {len(waveform_en)}, Framerate: {framerate_en}')
```

```python
def read_waveform(file_path):
    with wave.open(file_path, 'rb') as wf:
        n_frames = wf.getnframes()
        framerate = wf.getframerate()
        sample_width = wf.getsampwidth()
        n_channels = wf.getnchannels()
        raw_frames = wf.readframes(n_frames)

        # Convert byte data to a NumPy array, handling different sample widths
        if sample_width == 1:
            dtype = np.int8
        elif sample_width == 2:
            dtype = np.int16
        elif sample_width == 4:
            dtype = np.int32
        else:
            raise ValueError("Unsupported sample width")

        audio_data = np.frombuffer(raw_frames, dtype=dtype)

        # If stereo, separate channels (optional, for plotting individual channels)
        if n_channels == 2:
            audio_data = audio_data.reshape(-1, 2)
            # You can then plot audio_data[:, 0] for left channel and audio_data[:, 1] for right
        return audio_data, framerate
```

```python
# Generate time axis for plotting audio waveform
def get_time_axis(waveform, framerate):
    n_frames = len(waveform)
    return np.linspace(0, n_frames / framerate, num=n_frames)

# Example usage:
time_ar = get_time_axis(waveform_ar, framerate_ar)
time_en = get_time_axis(waveform_en, framerate_en)
```

```python
# Plot audio waveforms for Arabic and English files
plt.figure(figsize=(14, 6))
plt.subplot(2, 1, 1)
plt.plot(time_ar, waveform_ar if waveform_ar.ndim == 1 else waveform_ar[:, 0], color='tab:blue')
plt.title('Arabic Audio Waveform (solo3_ar)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(time_en, waveform_en if waveform_en.ndim == 1 else waveform_en[:, 0], color='tab:orange')
plt.title('English Audio Waveform (solo2_en)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Improved visualization: plot waveform and segment bars below for clarity, with vertical lines at segment ends spanning the full y-axis
import matplotlib.patches as mpatches

def plot_waveform_with_segment_bars(time, waveform, pyannote_segments, speechbrain_segments, title):
    fig, ax = plt.subplots(figsize=(16, 5), nrows=1)
    # Plot waveform (mono or left channel)
    ax.plot(time, waveform if waveform.ndim == 1 else waveform[:, 0], color='gray', alpha=0.7, label='Audio Waveform')

    # Precompute y-limits for consistent vertical lines
    y_min = -1.1 * np.max(np.abs(waveform))
    y_max = 1.1 * np.max(np.abs(waveform))

    # Plot pyannote segments as horizontal bars below waveform and vertical lines at segment ends
    for seg in pyannote_segments:
        ax.hlines(-0.8 * np.max(np.abs(waveform)), seg['start'], seg['end'], color='tab:blue', lw=8, label='Pyannote Segments')
        ax.vlines(seg['end'], y_min, y_max, color='tab:blue', linestyle='--', alpha=0.5, lw=1)
    # Plot speechbrain segments as horizontal bars below waveform and vertical lines at segment ends
    for seg in speechbrain_segments:
        ax.hlines(-0.95 * np.max(np.abs(waveform)), seg['start'], seg['end'], color='tab:orange', lw=8, label='SpeechBrain Segments')
        ax.vlines(seg['end'], y_min, y_max, color='tab:orange', linestyle='--', alpha=0.5, lw=1)

    # Only show one legend entry per label
    handles = [
        mpatches.Patch(color='tab:blue', label='Pyannote Segments'),
        mpatches.Patch(color='tab:orange', label='SpeechBrain Segments'),
        mpatches.Patch(color='gray', label='Audio Waveform')
    ]
    ax.legend(handles=handles, loc='upper right')
    ax.set_title(title)
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('Amplitude')
    ax.set_ylim([y_min, y_max])
    plt.tight_layout()
    plt.show()

# Plot for Arabic file
plot_waveform_with_segment_bars(time_ar, waveform_ar, pyannote_ar, speechbrain_ar, 'Arabic Audio with Pyannote & SpeechBrain Segments (solo3_ar)')
# Plot for English file
plot_waveform_with_segment_bars(time_en, waveform_en, pyannote_en, speechbrain_en, 'English Audio with Pyannote & SpeechBrain Segments (solo2_en)')
```

```python
# Plot 2-speaker audio waveform with pyannote and speechbrain segmentations overlaid (Arabic and English), using fixed colors for each system's speakers
import json
import matplotlib.patches as mpatches

# Fixed color mapping:
PYANNOTE_COLORS = ['tab:blue', 'tab:green']  # Speaker 1, Speaker 2 for pyannote
SPEECHBRAIN_COLORS = ['tab:orange', 'tab:red']  # Speaker 1, Speaker 2 for speechbrain

def add_speaker_field(segments, label_key='labels'):
    """Convert 'labels': ['Speaker 1'] to 'speaker': 0, ['Speaker 2'] to 1, etc."""
    for seg in segments:
        if label_key in seg and isinstance(seg[label_key], list) and len(seg[label_key]) > 0:
            label = seg[label_key][0]
            if '1' in label:
                seg['speaker'] = 0
            elif '2' in label:
                seg['speaker'] = 1
            else:
                seg['speaker'] = 0
        elif 'speaker' not in seg:
            seg['speaker'] = 0
    return segments

def plot_waveform_with_fixed_speaker_colors(time, waveform, pyannote_segments, speechbrain_segments, title):
    fig, ax = plt.subplots(figsize=(16, 5), nrows=1)
    # Plot waveform (mono or left channel)
    ax.plot(time, waveform if waveform.ndim == 1 else waveform[:, 0], color='gray', alpha=0.7, label='Audio Waveform')

    y_min = -1.1 * np.max(np.abs(waveform))
    y_max = 1.1 * np.max(np.abs(waveform))

    # Plot pyannote segments as horizontal bars below waveform and vertical lines at segment ends
    for seg in pyannote_segments:
        speaker = seg.get('speaker', 0)
        color = PYANNOTE_COLORS[speaker % 2]
        ax.hlines(-0.8 * np.max(np.abs(waveform)), seg['start'], seg['end'], color=color, lw=8, label=f'Pyannote Speaker {speaker+1}')
        ax.vlines(seg['end'], y_min, y_max, color=color, linestyle='--', alpha=0.5, lw=1)

    # Plot speechbrain segments as horizontal bars below waveform and vertical lines at segment ends
    for seg in speechbrain_segments:
        speaker = seg.get('speaker', 0)
        color = SPEECHBRAIN_COLORS[speaker % 2]
        ax.hlines(-0.95 * np.max(np.abs(waveform)), seg['start'], seg['end'], color=color, lw=8, label=f'SpeechBrain Speaker {speaker+1}')
        ax.vlines(seg['end'], y_min, y_max, color=color, linestyle=':', alpha=0.5, lw=1)

    # Build legend:
    handles = [
        mpatches.Patch(color='tab:blue', label='Pyannote Speaker 1'),
        mpatches.Patch(color='tab:green', label='Pyannote Speaker 2'),
        mpatches.Patch(color='tab:orange', label='SpeechBrain Speaker 1'),
        mpatches.Patch(color='tab:red', label='SpeechBrain Speaker 2'),
        mpatches.Patch(color='gray', label='Audio Waveform')
    ]
    ax.legend(handles=handles, loc='upper right', fontsize=10)
    ax.set_title(title)
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('Amplitude')
    ax.set_ylim([y_min, y_max])
    plt.tight_layout()
    plt.show()

# Load segmentation data for 2-speaker files
with open('../results/pyannote_predictions/two_speakers7_ar_pyannote.json') as f:
    pyannote_2sp_ar = json.load(f)
with open('../results/speechbrain_predictions/two_speakers7_ar_speechbrain.json') as f:
    speechbrain_2sp_ar = json.load(f)
with open('../results/pyannote_predictions/two_speakers7_en_pyannote.json') as f:
    pyannote_2sp_en = json.load(f)
with open('../results/speechbrain_predictions/two_speakers7_en_speechbrain.json') as f:
    speechbrain_2sp_en = json.load(f)

# Preprocess to add 'speaker' field from 'labels' if needed
pyannote_2sp_ar = add_speaker_field(pyannote_2sp_ar)
speechbrain_2sp_ar = add_speaker_field(speechbrain_2sp_ar)
pyannote_2sp_en = add_speaker_field(pyannote_2sp_en)
speechbrain_2sp_en = add_speaker_field(speechbrain_2sp_en)

# Plot for Arabic 2-speaker file
plot_waveform_with_fixed_speaker_colors(time_2sp_ar, waveform_2sp_ar, pyannote_2sp_ar, speechbrain_2sp_ar, 'Arabic 2-Speaker Audio with Pyannote (Blue/Green) & SpeechBrain (Orange/Red)')
# Plot for English 2-speaker file
plot_waveform_with_fixed_speaker_colors(time_2sp_en, waveform_2sp_en, pyannote_2sp_en, speechbrain_2sp_en, 'English 2-Speaker Audio with Pyannote (Blue/Green) & SpeechBrain (Orange/Red)')
```

## 7-Speaker Files: Quantitative Diarization Comparison

This section compares the diarization performance of pyannote and speechbrain on the 3-speaker Arabic and English files. We show Diarization Error Rate (DER) and the number of detected segments for each system and file.

```python
# Quantitative comparison for 3-speaker files: DER and segment count summary
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load evaluation/summary CSVs (already loaded earlier, but reload for clarity)
pyannote_eval = pd.read_csv('../results/Task 2 Results/pyannote_eval.csv')
speechbrain_eval = pd.read_csv('../results/Task 2 Results/speechbrain_eval.csv')
pyannote_summary = pd.read_csv('../results/Task 2 Results/pyannote_summary.csv')
speechbrain_summary = pd.read_csv('../results/Task 2 Results/speechbrain_summary.csv')

# File basenames for 3-speaker files
ar_file = 'three_speakers1_ar.wav'
en_file = 'three_speakers2_en.wav'

# Helper to get DER and n_segments for a file from a summary DataFrame
def get_metrics(df, file_basename):
    file_col = 'audio'  # Use the correct column name from your CSVs
    row = df[df[file_col].astype(str).str.contains(file_basename, na=False)]
    if not row.empty:
        der = row['DER'].values[0] if 'DER' in row.columns else None
        nseg = row['N_pred'].values[0] if 'N_pred' in row.columns else None
        return float(der) if der is not None else None, int(nseg) if nseg is not None else None
    else:
        return None, None

# Collect metrics
data = []
for sys_name, eval_df, sum_df in [
    ('pyannote', pyannote_eval, pyannote_summary),
    ('speechbrain', speechbrain_eval, speechbrain_summary)
]:
    for lang, fname in [('Arabic', ar_file), ('English', en_file)]:
        der, nseg = get_metrics(eval_df, fname)
        data.append({'System': sys_name, 'Language': lang, 'DER': der, 'Segments': nseg})

metrics_df = pd.DataFrame(data)

# Display table
print('DER and Segment Count for 3-Speaker Files:')
display(metrics_df)

# Bar plot for DER
plt.figure(figsize=(7,4))
sns.barplot(data=metrics_df, x='Language', y='DER', hue='System', palette='Set1')
plt.title('Diarization Error Rate (DER) for 3-Speaker Files')
plt.ylabel('DER')
plt.show()

# Bar plot for segment count
plt.figure(figsize=(7,4))
sns.barplot(data=metrics_df, x='Language', y='Segments', hue='System', palette='Set2')
plt.title('Number of Segments Detected for 3-Speaker Files')
plt.ylabel('Number of Segments')
plt.show()
```

## 8- Speaker Overlap Analysis

This section analyzes how well each diarization system handles overlapping speech segments. We will compare the total duration and proportion of overlapping speech detected by pyannote and speechbrain, and visualize the results.

```python
# Analyze and visualize speaker overlap handling for pyannote and speechbrain
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os, json, re

def compute_overlap_duration(segments):
    segs = sorted([(s['start'], s['end']) for s in segments], key=lambda x: x[0])
    if not segs:
        return 0.0
    overlap = 0.0
    prev_end = segs[0][1]
    for i in range(1, len(segs)):
        start, end = segs[i]
        if start < prev_end:
            overlap += min(end, prev_end) - start
        prev_end = max(prev_end, end)
    return overlap

# List all prediction files for both systems
pyannote_dir = '../results/pyannote_predictions/'
speechbrain_dir = '../results/speechbrain_predictions/'
pyannote_files = [f for f in os.listdir(pyannote_dir) if f.endswith('.json')] if os.path.exists(pyannote_dir) else []
speechbrain_files = [f for f in os.listdir(speechbrain_dir) if f.endswith('.json')] if os.path.exists(speechbrain_dir) else []

# Build mapping from base name to file for each system
def get_base_name(f, system):
    if system == 'pyannote':
        return re.sub(r'_pyannote\.json$', '', f)
    else:
        return re.sub(r'_speechbrain\.json$', '', f)
pyannote_map = {get_base_name(f, 'pyannote'): f for f in pyannote_files}
speechbrain_map = {get_base_name(f, 'speechbrain'): f for f in speechbrain_files}

# Find common base names
common_bases = set(pyannote_map.keys()) & set(speechbrain_map.keys())
print('Common audio files:', sorted(common_bases))

# Compute overlap for each file and system
results = []
for base in sorted(common_bases):
    py_file = pyannote_map[base]
    sb_file = speechbrain_map[base]
    with open(os.path.join(pyannote_dir, py_file)) as pf:
        pyannote_segs = json.load(pf)
    with open(os.path.join(speechbrain_dir, sb_file)) as sf:
        speechbrain_segs = json.load(sf)
    all_times = [(s['start'], s['end']) for s in pyannote_segs + speechbrain_segs]
    total_duration = max([end for _, end in all_times]) - min([start for start, _ in all_times]) if all_times else 0.0
    py_overlap = compute_overlap_duration(pyannote_segs)
    sb_overlap = compute_overlap_duration(speechbrain_segs)
    results.append({'file': base, 'system': 'pyannote', 'overlap_sec': py_overlap, 'total_sec': total_duration, 'overlap_pct': py_overlap/total_duration if total_duration else 0})
    results.append({'file': base, 'system': 'speechbrain', 'overlap_sec': sb_overlap, 'total_sec': total_duration, 'overlap_pct': sb_overlap/total_duration if total_duration else 0})

overlap_df = pd.DataFrame(results)
print('overlap_df columns:', overlap_df.columns.tolist())
print(overlap_df.head())

if not overlap_df.empty:
    plt.figure(figsize=(10,5))
    sns.barplot(data=overlap_df, x='file', y='overlap_pct', hue='system')
    plt.ylabel('Overlap Proportion')
    plt.title('Proportion of Overlapping Speech Detected by System')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print('No overlap data to plot.')
```

```python
# Diarization Error Rate (DER) comparison using evaluation CSVs
import pandas as pd
import matplotlib.pyplot as plt

# Load evaluation CSVs
pyannote_eval = pd.read_csv('../results/Task 2 Results/pyannote_eval.csv')
speechbrain_eval = pd.read_csv('../results/Task 2 Results/speechbrain_eval.csv')

# Ensure the audio file column is named 'audio' and DER column is 'DER'
pyannote_eval = pyannote_eval.rename(columns={col: col.lower() for col in pyannote_eval.columns})
speechbrain_eval = speechbrain_eval.rename(columns={col: col.lower() for col in speechbrain_eval.columns})

# Merge on audio file name (base name, without extension)
pyannote_eval['base'] = pyannote_eval['audio'].apply(lambda p: os.path.splitext(os.path.basename(str(p)))[0])
speechbrain_eval['base'] = speechbrain_eval['audio'].apply(lambda p: os.path.splitext(os.path.basename(str(p)))[0])

merged = pd.merge(pyannote_eval[['base','der']], speechbrain_eval[['base','der']], on='base', suffixes=('_pyannote','_speechbrain'))

rows = []
for idx, row in merged.iterrows():
    if row['der_pyannote'] < row['der_speechbrain']:
        better = 'pyannote'
    elif row['der_speechbrain'] < row['der_pyannote']:
        better = 'speechbrain'
    else:
        better = 'tie'
    rows.append({'file': row['base'], 'pyannote_DER': row['der_pyannote'], 'speechbrain_DER': row['der_speechbrain'], 'better_model': better})

compare_df = pd.DataFrame(rows)
display(compare_df)

# Summary bar plot
summary = compare_df['better_model'].value_counts().reset_index()
summary.columns = ['Model', 'Count']
plt.figure(figsize=(5,3))
plt.bar(summary['Model'], summary['Count'], color=['#1f77b4', '#ff7f0e', '#888888'])
plt.title('Which Model Has Lower DER (Eval CSV)')
plt.ylabel('Number of Files')
plt.show()

# Print overall winner
winner = summary.loc[summary['Count'].idxmax(), 'Model']
print(f'Overall, the model with lower DER (better diarization) is: {winner.upper()}')
```
